```julia
using DataFrames
using Statistics
using CSV
using Plots
```

```julia
input="results.csv"
```

```julia
mydata=CSV.read(input, delim=",")
```

```julia
vvdev = by(mydata, :size, :3 => std)
vvmean = by(mydata, :size, :3 => mean)
vmdev = by(mydata, :size, :4 => std)
vmmean = by(mydata, :size, :4 => mean)
```

```julia
vm = DataFrame(size=vmdev[:1],mean=vmmean[:2],dev=vmdev[:2],flag=zeros(100))
```

```julia
vv = DataFrame(size=vvdev[:1],mean=vvmean[:2],dev=vvdev[:2],flag=ones(100))
```

```julia
df = append!(vv, vm)
```

```julia
scatter(vmmean[:1], vmmean[:2], yerr=vmdev[:2])
```

```julia
scatter(vvmean[:1], vvmean[:2], yerr=vvdev[:2])
```

```julia
scatter(df[:size], df[:mean], group=df[:flag], layout=2, yerr=df[:dev])
```
